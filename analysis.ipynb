## Plotings 

```python


# Use common column names (case-insensitive)
cols_lower = {c.lower(): c for c in df.columns}
age_col = cols_lower.get("age", "Age") if "age" in cols_lower else df.columns[0]
height_col = cols_lower.get("height", "Height") if "height" in cols_lower else df.columns[1]

fig, axes = plt.subplots(1, 2, figsize=(12, 4), constrained_layout=True)

axes[0].hist(df[age_col].dropna(), bins=20, edgecolor="white")
axes[0].set_title(f"Distribution of {age_col}")
axes[0].set_xlabel(age_col)
axes[0].set_ylabel("Count")

axes[1].hist(df[height_col].dropna(), bins=20, edgecolor="white")
axes[1].set_title(f"Distribution of {height_col}")
axes[1].set_xlabel(height_col)
axes[1].set_ylabel("Count")

plt.show()
```
